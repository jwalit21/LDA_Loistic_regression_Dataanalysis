```python
#logistic regression
```

```python
from sklearn.linear_model import LogisticRegression


logireg_mul = LogisticRegression()


x = data3.loc[ : , data3.columns != 'Sold' ]
y = data3['Sold']


logireg_mul.fit(x,y)
```

```python
logireg_mul.predict_proba(x)
```

```python
logireg_mul.predict(x)
```

```python
y_pred_03 = ( logireg_mul.predict_proba(x)[:,1] >= 0.3 ).astype(bool)
```

```python
y_pred_02 = ( logireg_mul.predict_proba(x)[:,1] >= 0.2 ).astype(bool)
```

```python
from sklearn.metrics import confusion_matrix
confusion_matrix(y,y_pred_03)
```

```python
confusion_matrix(y,y_pred_02)
```

```python
from sklearn.metrics import precision_score , recall_score , roc_auc_score
```

```python
precision_score(y,y_pred_03)
```

```python
recall_score(y,y_pred_03)
```

```python
roc_auc_score(y,y_pred_03)
```
